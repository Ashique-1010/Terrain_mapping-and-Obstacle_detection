```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/Ashique-1010/Datasets-for-EDA-ML/main/commercial_lidar.csv"
data = pd.read_csv(url)
data
```

```python
data.columns
```

```python
cols = ['time', 'r1', 'r2', 'r3', 'r4', 'ds1', 'ds2', 'ds3', 'ds4']
data.columns = cols
data
```

```python
data.info()
data.describe()
```

```python
# LiDAR sensor head origin location
origin = np.array([0.160, -3.13, -1.22])

# Line of sight vectors for each beam
beams = {
    'Beam1': np.array([-0.2998, -0.3996, -0.8663]),
    'Beam2': np.array([0.0810, -0.4900, -0.8680]),
    'Beam3': np.array([0.1178, -0.7326, -0.6703]),
    'Beam4': np.array([0.4192, -0.2778, -0.8643])
}

# Range data for each beam

range_data = data[['r1', 'r2', 'r3', 'r4']].values
range_data
```

```python
points = []

for i, (beam_name, direction) in enumerate(beams.items()):
    r = range_data[:, i]

    for range_value in r:
        point = origin + range_value * direction            # Calculate the points in 3D space
        points.append(point)

points = np.array(points)
points
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], marker='o')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Point Cloud from LiDAR Data')
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=1.0, min_samples=5)
labels = dbscan.fit_predict(points)

# Plotting the clustered point cloud
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for k, col in zip(unique_labels, colors):
    if k == -1:
        col = [0, 0, 0, 1]  # Black for noise

    class_member_mask = (labels == k)
    xyz = points[class_member_mask]
    ax.scatter(xyz[:, 0], xyz[:, 1], xyz[:, 2], c=[col], marker='o')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Point Cloud with DBSCAN Clustering')
plt.show()
```

```python

# Improved plotting with distinct colors for each cluster and noise
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Unique labels (clusters and noise)
unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))

# Plot each cluster with a different color
for k, col in zip(unique_labels, colors):
    if k == -1:
        col = [0, 0, 0, 1]  # Black color for noise

    class_member_mask = (labels == k)
    xyz = points[class_member_mask]
    ax.scatter(xyz[:, 0], xyz[:, 1], xyz[:, 2], c=[col], marker='o', label=f'Cluster {k}' if k != -1 else 'Noise')

ax.legend()
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Point Cloud with DBSCAN Clustering')
plt.show()
```
